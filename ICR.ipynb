This competition really want us to experiment on modelling side more than we think. Listing down below the common model ensembling techniques that can be utilised to get better score and rank

## 1. Voting Ensemble:
Hard Voting: In hard voting, each model in the ensemble predicts a class label, and the final prediction is determined by majority voting.
Soft Voting: In soft voting, each model in the ensemble predicts the probabilities for each class, and the final prediction is based on averaging the predicted probabilities across models.

## 2. Weighted Average:
In weighted average ensemble, each model's prediction is weighted based on its performance or expertise. Models with better performance or higher confidence are given higher weights when combining their predictions.

## 3. Stacking:
Stacking is a meta-ensemble technique where the predictions of individual models are used as input features for a meta-model (also known as a meta-learner).
The individual models are trained on the training data, and their predictions are then used as features along with the original features to train the meta-model.
The meta-model learns how to combine the predictions of the individual models to make the final prediction.

## 4. Bagging (Bootstrap Aggregating):
Bagging involves training multiple models on different subsets of the training data, typically created through bootstrapping (random sampling with replacement).
Each model is trained independently, and the final prediction is made by aggregating the predictions of all models, either by majority voting (classification) or averaging (regression).
Random Forest is a popular example of a bagging ensemble that combines decision trees.

## 5. Boosting:
Boosting is an iterative ensemble technique where models are trained sequentially to correct the mistakes made by previous models.
Each model is trained on a modified version of the training data, where more weight is given to misclassified samples.
Boosting algorithms such as AdaBoost, Gradient Boosting Machines (GBM), XGBoost, CatBoost, and LightGBM are widely used.

## 6. Blending:
Blending involves training multiple models on the training data and holding out a separate validation set for evaluation.
The predictions of individual models on the validation set are combined using a simple technique such as averaging or majority voting.
The blended predictions can be used as the final prediction or as features for another model.

## 7. Meta-Stacking:
Meta-stacking is an extension of stacking where multiple levels of stacking are used.
In addition to using individual models' predictions as input features for the meta-model, predictions from intermediate models can also be used as input features for subsequent levels of stacking.
These ensemble techniques provide various ways to combine the predictions of GBM, CatBoost, and other machine learning models to improve overall performance. The choice of ensemble technique depends on the characteristics of the models, the nature of the problem, and the available data. It is recommended to experiment with different ensemble techniques to find the most effective approach for a specific task.

```python
from sklearn.metrics import log_loss

def balanced_log_loss(y_true, y_pred):
    # y_true: correct labels 0, 1
    # y_pred: predicted probabilities of class=1
    # calculate the number of observations for each class
    N_0 = np.sum(1 - y_true)
    N_1 = np.sum(y_true)

    # calculate the weights for each class to balance classes
    w_0 = 1 / N_0
    w_1 = 1 / N_1

    # calculate the predicted probabilities for each class
    p_1 = np.clip(y_pred, 1e-15, 1 - 1e-15)
    p_0 = 1 - p_1

    # calculate the summed log loss for each class
    log_loss_0 = -np.sum((1 - y_true) * np.log(p_0))
    log_loss_1 = -np.sum(y_true * np.log(p_1))

    # calculate the weighted summed logarithmic loss
    # (factgor of 2 included to give same result as LL with balanced input)
    balanced_log_loss = 2*(w_0 * log_loss_0 + w_1 * log_loss_1) / (w_0 + w_1)

    # return the average log loss
    return balanced_log_loss/(N_0+N_1)
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# 로지스틱 회귀와 KNN을 기반으로 소프트 보팅 방식으로 보팅 분류기 만들기

# 개별 모델은 로지스틱 회귀와 KNN임
lr_clf = LogisticRegression()
knn_clf = KNeighborsClassifier(n_neighbors=3)
rf_clf = RandomForestClassifier(random_state=777)

# 개별 모델을 소프트 보팅 기반의 앙상블 모델로 구현한 분류기
vo_clf = VotingClassifier(estimators=[('LR', lr_clf), ('KNN', knn_clf), ('RF', rf_clf)], voting='soft')

# VotingClassifier 학습/예측/평가
vo_clf.fit(x_train_stan, y_train)
pred = vo_clf.predict(x_test_stan)
print('Voting 분류기 정확도: {0:.4f}'.format(accuracy_score(y_test, pred)))

# 개별 모델의 학습/예측/평가
classifiers = [lr_clf, knn_clf, rf_clf]
for classifier in classifiers:
    classifier.fit(x_train_stan, y_train)
    pred = classifier.predict(x_test_stan)
    class_name = classifier.__class__.__name__
    print('{0} 정확도: {1:.4f}'.format(class_name, accuracy_score(y_test,pred)))
```
