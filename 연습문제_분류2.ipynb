# 1. Lightgbm

* 데이터 '와인품질분류.zip'입니다.

1. 필요한 라이브러리를 가져옵니다.

```python
import pandas as pd
import lightgbm as lgbm
```

2. 데이터를 불러옵니다.

```python
train = pd.read_csv('/content/sample_data/train.csv')
test = pd.read_csv('/content/sample_data/test.csv')
smpl_sub = pd.read_csv('/content/sample_data/sample_submission.csv')
```

```python
train.head()
```

```python
test.head()
```

3. 변수를 변환합니다. 

```python
# type에는 white와 red 두 종류가 있습니다.
# 각각 0,1로 변환합니다.
size_mapping = {'white':0, 'red':1}
train['type']=train['type'].map(size_mapping)
test['type']=test['type'].map(size_mapping)
```

```python
train.head()
```

```python
test.head()
```

```python
# 모델에 입력하기 전 데이터를 정형합니다.
# train_x : train의 'index', 'quality' 피처를 제거하고 저장합니다.
# train_y : train의 'quality'를 저장합니다.
# test_x : test의 'index'를 제거하고 저장합니다.

train_x = train.drop(['index','quality'],axis=1)
train_y = train['quality']
test_x = test.drop('index',axis=1)
```

```python
train_x.shape, train_y.shape, test_x.shape
```

```python
train_x.head()
```

```python
train_y.head()
```

```python
test_x.head()
```

4. LightGBM 모델을 생성 및 훈련합니다.

```python
from lightgbm import LGBMClassifier

import pandas as pd
import numpy as np
lgbm = LGBMClassifier()
lgbm.fit(train_x, train_y)
```

5. 생성한 모델로 예측 데이터를 생성합니다.

```python
y_pred = lgbm.predict(test_x)
```

6. 제출파일을 생성합니다.

```python
smpl_sub.head()
```

```python
#서브미션 파일의 quality 변수에 예측값을 대입합니다.
smpl_sub['quality'] = y_pred
```

```python
smpl_sub.head()
```

```python
# csv 파일로 저장합니다.
smpl_sub.to_csv('submission.csv', index=False)
```

# 2. XGBoost

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from xgboost import XGBClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
test.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
train_one = pd.get_dummies(train)
test_one = pd.get_dummies(test)
```

```python
# 모델 정의
model = XGBClassifier()
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수
X = train_one.drop('quality',axis=1)
y = train_one['quality']

# fit 메소드를 이용해 모델 학습
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
model.fit(X,y)
```

```python
# predict 메소드와 test_one 데이터를 이용해 품질 예측
y_pred = model.predict(test_one)
```

```python
# sample_submission.csv 파일을 불러와 예측된 값으로 채워 주기
sub = pd.read_csv('/content/submission.csv')
sub['quality'] = y_pred + 3
sub.head()
```

```python
# 정답파일 내보내기
sub.to_csv('xgb_pred.csv',index = False)
```

```python
# 배운점: 최근 xgboost 버전은 실행했을 때 'Invalid classes inferred from unique values of `y`.  Expected: [0 1 2 3 4 5 6], got [3 4 5 6 7 8 9]'라는 오류가 뜨는 경우가 있는데 이는
#label을 새로 encoding 한 뒤 다시 학습 코드를 실행하여 해결할 수 있었음. 또한 파일을 내보내는 것은 'to_csv'사용하면 가능하다는 것을 알게 되었음.
```

# 3. 보팅

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
train_one = pd.get_dummies(train)
test_one = pd.get_dummies(test)
```

```python
# 모델 정의 (LightGBM, XGBoost, Random Forest)
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
LGBM = LGBMClassifier()
XGB = XGBClassifier()
RFC = RandomForestClassifier()
# VotingClassifier 정의
from sklearn.ensemble import VotingClassifier
VC = VotingClassifier(estimators=[('lgbm',LGBM),('xgb',XGB),('rfc',RFC)], voting='soft')
```

```python
# 모델 학습

# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수
X = train_one.drop('quality',axis=1)
y = train_one['quality']

# fit 메소드를 이용해 모델 학습
VC.fit(X,y)
```

```python
# predict 메소드와 test_one 데이터를 이용해 품질 예측
y_pred = VC.predict(test_one)
y_pred
```
